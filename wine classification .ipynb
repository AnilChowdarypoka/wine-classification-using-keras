**Feature Description**


**Alcohol**: Wine's ethanol content, influencing flavor, body, and perceived warmth.

**Malic acid:** Tartness-contributing organic acid in grapes and wine.

**Ash:** Residue after incineration, reflects soil influence on wine composition.

**Alcalinity of ash:** Measures alkaline ash components affecting wine's pH and acidity.

**Magnesium:** Mineral element in wine, influencing taste and character.

**Total phenols:** A group of compounds impacting color, flavor, and aging potential.

**Flavanoids:** Phenolic compounds shaping wine color, flavor, and health benefits.

**Nonflavanoid phenols:** Wine phenolics contributing to taste and mouthfeel.

**Proanthocyanins:** Flavonoids adding astringency, bitterness, and color stability to red wine.

**Color intensity:** Depth and richness of wine color, revealing grape variety and winemaking.

**Hue:** Wine's actual color, part of its overall visual profile.

**OD280/OD315 of diluted wines:** Measures optical density and clarity in wine.

**Proline:** An amino acid found in grapes and wine, impacting taste.

```python

def replace_outliers_with_mean(data):
  for column in data.columns:
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    data[column] = data[column].apply(lambda x: data[column].mean() if x < lower_bound or x > upper_bound else x)
  return data

data = replace_outliers_with_mean(data)
```

```python
encoder = LabelEncoder()
data['target'] = encoder.fit_transform(data[['target']])


```

```python
X_train, X_test, y_train, y_test = train_test_split( data.drop('target' , axis = 1), data['target'], test_size=0.2, random_state=42)
```

```python
scaler = StandardScaler().fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)

y_train = to_categorical(y_train, num_classes=3)
y_test = to_categorical(y_test, num_classes=3)
```

```python
accuracies , errors = {} , {}
```

```python

def initilize_model (data_shape , optimizer):
    model = Sequential()
    model.add(Dense(128, activation='relu', input_dim = data_shape))
    model.add(Dense(64, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(32, activation='relu'))
    model.add(Dense(3, activation='softmax'))
    model.compile(loss='categorical_crossentropy', optimizer=optimizer, metrics=['accuracy'])
    return model
```

```python
adam_model = initilize_model(X_train.shape[1], Adam(learning_rate=0.001))
adam_history = adam_model.fit(X_train, y_train,epochs=20, batch_size=1, verbose=1 ,  validation_data=(X_test, y_test))
test_loss, test_accuracy = adam_model.evaluate(X_test, y_test)
print("-----------------------------------------")
adam_model.summary()
print("-----------------------------------------")
accuracies['Adam with 0.001 learning rate'] = test_accuracy
print(f"Test accuracy for  Adam with 0.001 learning rate is : {test_accuracy}")
```

```python
sgd_model = initilize_model(X_train.shape[1], SGD(learning_rate=0.001))
sgd_history = sgd_model.fit(X_train, y_train,epochs=20, batch_size=1, verbose=1 , validation_data=(X_test, y_test))
test_loss, test_accuracy = sgd_model.evaluate(X_test, y_test)
print("-----------------------------------------")
sgd_model.summary()
print("-----------------------------------------")
accuracies['SGD with 0.001 learning rate'] = test_accuracy

print(f"Test accuracy for  SGD with 0.001 learning rate is : {test_accuracy}")
```

```python
def plot_graphs(history , optimizer):

    loss = history.history['loss']
    val_loss = history.history['val_loss']
    accuracy = history.history['accuracy']
    val_accuracy = history.history['val_accuracy']
    sns.set(style='whitegrid')

    fig, axes = plt.subplots(1, 2, figsize=(12, 5))

    ax1 = axes[0]
    ax1.plot(range(1, len(loss) + 1), loss, label='Training Loss')
    ax1.plot(range(1, len(val_loss) + 1), val_loss, label='Validation Loss')

    ax1.set_title(f'Training and Validation Loss for {optimizer}')
    ax1.set_xlabel('Epochs')
    ax1.set_ylabel('Loss')
    ax1.legend()

    ax2 = axes[1]
    ax2.plot(range(1, len(accuracy) + 1), accuracy, label='Training Accuracy')
    ax2.plot(range(1, len(val_accuracy) + 1), val_accuracy, label='Validation Accuracy')

    ax2.set_title(f'Training and Validation Accuracy for {optimizer}')
    ax2.set_xlabel('Epochs')
    ax2.set_ylabel('Accuracy')
    ax2.legend()

    plt.tight_layout()
    plt.show()

```

```python
plot_graphs(adam_history , 'Adam')
```

```python
plot_graphs(sgd_history , 'SGD')
```

```python

def plot_confusion_matrix(model, X_test , y_test  ):
  predicted_values = model.predict(X_test)
  conf_matrix = confusion_matrix(np.argmax(y_test, axis=1), np.argmax(predicted_values, axis=1))
  sns.heatmap(conf_matrix, annot=True, cmap='YlGnBu', fmt='d', xticklabels=[ 1, 2 , 3], yticklabels=[ 1, 2 , 3])
  plt.xlabel('Predicted')
  plt.ylabel('True')
  plt.show()
```

```python
plot_confusion_matrix(sgd_model , X_test , y_test )
```
